- Check if a Number is Prime.
  - Imput a number and check if it is prime or not.
  - Input: 7 ----> Output: "Prime"

```python
num = int(input('Enter a number:'))
i = 2
while i<=num-1:
    if num%i==0:
        print(f"{num} is not a prime number")
        break
    i=i+1
else:
    print(f"{num} is a prime number")
```

- Check BMI (BODY MASS Index)
  - Based on BMI value, categorize as "UnderWeight", "Normal", "OverWeight", "Obese"
  - Input: 24 ---> Output: "Normal"

```python
weight = float(input("Enter weight in kg: "))
height = float(input("Enter height in meters: "))

bmi = weight / (height * height)

print("BMI:", round(bmi, 2))

if bmi < 18.5:
    print("UnderWeight")
elif bmi < 25:
    print("Normal")
elif bmi < 30:
    print("OverWeight")
else:
    print("Obese")
```

- Convert Temperature (Celsius to Fahrenheit/Kelvin)
  - Take Celsius input and allow users to convert to Fahrenheit or Kelvin.
  - Input: 100C , "F" ---> Output: 212 F

```python
celsius = float(input('Enter the temperature in celsius:'))
choice = input('convert to Fahrenheit(F) or Kelvin(K)')
if choice == 'F':
    fahrenheit = (celsius * 9/5) + 32
    print(f'{celsius} c is equal to {fahrenheit} f')
elif choice == 'K':
    kelvin = celsius + 273.15
    print(f'{celsius} c is equal to {kelvin} k')
else:
    print('Invalid choice')    
```

- Employee Bonus Based on Experience
  - If experience > 5 years, give a 20% bonus.
  - If experience is 3-5 years, give 10% bonus.
