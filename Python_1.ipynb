### 4) Write a Python program which accepts a sequence of comma-separated numbers from user and generate a list and a tuple with those numbers

- **The ***split()*** method splits a string into a list.**
- **You can specify the separator; default separator is any whitespace**

```python
values = input('Enter your comma separated numbers:')
list1 = values.split(',')
tuple1 = tuple(list1)
print("Your list :",list1)
print("Your tuple :",tuple1)
```

- Other example

```python
text = "Hello, My name is Prajwal Ghotkar, I am from USA"
list1 = text.split(",")
tuple1 = tuple(list1)
print('Your new generated list:',list1)
print('Your new generated tuple:',tuple1)
```

```python
text = "apple#banana#cherry"
list1 = text.split('#')
print("Your new list:",list1)
```

### 7) Write a Python program that accepts an interger (n) and computes the value of n+nn+nnn

```python
a = int(input('Enter the number:'))
n1 = int("%s" %a)
n2 = int("%s%s" %(a,a))
n3 = int("%s%s%s" %(a,a,a))
print(n1)
print(n2)
print(n3)
print(n1+n2+n3)
```
