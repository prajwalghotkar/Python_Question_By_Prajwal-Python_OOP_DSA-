- **Write a python script to calculate area of rectangle.**

```python
length = float(input('Enter length:'))
width = float(input('Enter width:'))
area = length * width
print('Area of rectangle is:',area)
```

- **Write a python program to calculate simple interest.**

```python
principal = float(input('Enter principal amount:'))
rate = float(input('Enter rate of interest:'))
time = float(input('Enter Time (in year):'))
simple_interest = (principal*rate*time)/100
print('simple interest is:',simple_interest)
```

- **Write a python script to swap data of two variables.**

```python
a = int(input('Enter a 1st number:'))
b = int(input('Enter a 2nd number:'))
a,b = b,a
print('After swapping:')
print('a =',a)
print('b =',b)
```

```python
a = int(input('Enter a 1st number:'))
b = int(input('Enter a 2nd number:'))

temp = a
a = b
b = temp
print('After swapping two number:')
print('a =',a)
print('b =',b)
```

- **Write a python script to print two given words in dictionary order.**

```python
word1 = input('Enter 1st word:')
word2 = input('Enter 2nd word:')
if word1 < word2:
    print(word1,word2)
else:
    print(word2,word1)
```

```python
word1 = input('Enter 1st word:')
word2 = input('Enter 2nd word:')
if word1.lower() < word2.lower():
    print(word1,word2)
else:
    print(word2,word1)
```

- **Write a python script to check whetjer a given number is a three digit number or not.**

```python
num = int(input('Enter a number:'))
if 100 <= abs(num) <= 999:
    print(f'{num} is a three digit number')
else:
    print(f'{num} is not a three digit number')
```

```python
num = int(input("Enter a number: "))
if (100 <= num <= 999) or (-999 <= num <= -100):
    print("It is a three-digit number")
else:
    print("It is NOT a three-digit number")
```

- **Write a python script to check whether a given year is a leap year or not.**

```python
year = int(input('Enter year:'))
if (year%4==0 and year%100!=0) or year%400==0:
    print(f'{year} is a leap year')
else:
    print(f'{year} is not leap year')
```

- **Write a python script to check whetjer a given number is positive,negative or zero.**

```python
num = float(input("Enter a number: "))
if num > 0:
    print("The number is Positive")
elif num < 0:
    print("The number is Negative")
else:
    print("The number is Zero")
```

- **Write a python script to accept one complex number from the user and display the greater number between real part and imaginary part.**

```python
n = complex(input('Enter your number in the form of "1+3j":'))
real_part = n.real
imag_part = n.imag
if real_part > imag_part:
    print("Real part is greater:", real_part)
elif imag_part > real_part:
    print("Imaginary part is greater:", imag_part)
else:
    print("Both parts are equal:", real_part)
```
